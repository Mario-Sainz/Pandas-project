```python
import pandas as pd
```

```python
data = pd.read_csv("../data/attacks.csv",encoding='cp1252')
```

```python
data.head()
```

```python
data
```

```python
data.info()
```

```python
# Elimino las columnas que veo que prácticamente estan formadas por datos nulos.
data1 = data.drop(['Unnamed: 23', 'Unnamed: 22', 'Case Number'], axis=1)
```

```python
data1
```

```python
# Elimino las filas que contienen únicamente valores nulos.
data2 = data1.dropna(how = "all")
```

```python
data2
```

```python
# Elimino, una vez pensada mi hipótesis, las columnas que no me aportan nada.
data3 = data2.drop(['Name', 'Fatal (Y/N)', 'Time', 'Investigator or Source', 'pdf', 'href formula', 'href', 'Case Number.1', 'Case Number.2', 'original order'  ], axis=1)
```

```python
data3
```

```python
# Cambio "NaN" por guiones
data4 = data3.fillna('-')
```

```python
data4
```

```python
# Renombrar las columnas que tenían un espacio y no detectaba bien
data5 = data4.rename(columns = {'Sex ': 'Sex', 'Species ': 'Species'}, inplace = False)
```

```python
data5
```

```python
data5.to_csv('data5.csv')
```
